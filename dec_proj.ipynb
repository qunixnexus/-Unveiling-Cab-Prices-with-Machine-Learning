ELIMINATING OUTLIERS

remove_outliers(df1, col):
 - Calculates the quartiles (Q1, Q2, Q3) and interquartile range (IQR) for a specified column.
 - Determines lower and upper whiskers to identify outliers.
 - Clips values outside the whiskers to the whisker values, effectively removing outliers.
 - Returns the modified DataFrame.

treat_all_outliers(df1, col_list):
 - Iterates through a list of column names (col_list).
 - Calls remove_outliers for each column, treating outliers in each specified column.
 - Returns the DataFrame with outliers treated.

```python
def remove_outliers(df1, col):
    Q1 = df1[col].quantile(0.25)
    Q2 = df1[col].quantile(0.50)
    Q3 = df1[col].quantile(0.75)
    
    IQR = Q3 - Q1
    
    lower_whisker = Q1 - 1.5*IQR
    upper_whisker = Q3 + 1.5*IQR
    
    print("----------------------------------------------------")
    print("col = ", col, "Q1 = ", Q1, "Q2 = ", Q2, "Q3 = ", Q3)
    print("----------------------------------------------------")
    df1[col] = np.clip(df1[col], lower_whisker, upper_whisker)
    
    return df1

def treat_all_outliers(df1, col_list):
    print("col_list:", col_list)
    
    for c in col_list:
        df1 = remove_outliers(df1, c)
        
    return df1
```

Calculating distance between coordinates

Haversine Distance Calculation:
 - The code uses the Haversine formula to calculate the distance between two geographic coordinates, specifically the pickup and dropoff locations.
 - For each row in the DataFrame, it extracts the pickup and dropoff longitude and latitude values.
 - These coordinates are then converted into tuples (loc1 and loc2) representing the two locations.
 - The Haversine distance (c) is calculated using the haversine function from the haversine library, measuring the distance between the two locations.
 - The calculated distances are appended to the list travel_dist.

```python
import haversine as hs

travel_dist = []

for pos in range(len(df['pickup_longitude'])):
    long1, lat1, long2, lat2 = [df['pickup_longitude'][pos], df['pickup_latitude'][pos], df['dropoff_longitude'][pos], df['dropoff_latitude'][pos]]
    loc1 = (lat1, long1)
    loc2 = (lat2, long2)
    
    c = hs.haversine(loc1, loc2)
    travel_dist.append(c)
    
print(travel_dist)
df['dist_travel'] = travel_dist
```

Metrics evaluation

Mean Absolute Error (MAE):
Definition: The mean absolute difference between the predicted and actual values.
Interpretation: The average absolute error between predictions and true values.

Mean Squared Error (MSE):
Definition: The mean of the squared differences between predicted and actual values. 
Interpretation: MSE penalizes larger errors more heavily than MAE.

R-squared (Coefficient of Determination):
Definition: The proportion of the variance in the dependent variable that is predictable from the independent variables.
Interpretation: R-squared ranges from 0 to 1; a higher value indicates a better fit. It measures the proportion of variability in the dependent variable that is explained by the model.

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"MAE: {mae:.4f}")
print(f"MSE: {mse:.4f}")
print(f"R-squared: {r2:.4f}")
```

Random Forest Regression:
 - Random Forest Regression is an ensemble machine learning algorithm that extends the idea of decision tree regression. It builds multiple decision trees during training and merges their predictions to improve accuracy and reduce overfitting.
 - Random Forest Regression is a powerful and flexible algorithm suitable for a variety of regression tasks, especially when dealing with complex datasets with multiple features. 
 - It tends to provide robust and accurate predictions while mitigating issues like overfitting.

Use of the RandomForestRegressor from scikit-learn to train a Random Forest regression model and make predictions on test data.
